```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos del archivo Excel
archivo_excel = "/content/shopsales23.xlsx"
datos = pd.read_excel(archivo_excel)

# Contar la frecuencia de cada producto
productos_mas_vendidos = datos['lineitem_id'].value_counts()

# Obtener el producto m√°s vendido
producto_mas_vendido = productos_mas_vendidos.idxmax()
cantidad_mas_vendida = productos_mas_vendidos.max()

# Crear la gr√°fica
plt.figure(figsize=(10, 6))
productos_mas_vendidos.plot(kind='bar')
plt.title('Productos m√°s vendidos')
plt.xlabel('ID del producto')
plt.ylabel('N√∫mero de ventas')
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el producto m√°s vendido
print(f"El producto m√°s vendido es el ID {producto_mas_vendido} con {cantidad_mas_vendida} ventas.")

# Mostrar la gr√°fica
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos del archivo Excel
archivo_excel = "/content/shopsales23.xlsx"
datos = pd.read_excel(archivo_excel)

# Convertir la columna de fechas a tipo datetime
datos['fulfilled_at'] = pd.to_datetime(datos['fulfilled_at'])

# Extraer el mes de cada fecha
datos['mes'] = datos['fulfilled_at'].dt.month

# Contar la frecuencia de ventas por mes
ventas_por_mes = datos['mes'].value_counts()

# Obtener el mes con m√°s ventas
mes_mas_ventas = ventas_por_mes.idxmax()
ventas_mes_mas_ventas = ventas_por_mes.max()

# Crear la gr√°fica de pastel
plt.figure(figsize=(8, 8))
plt.pie(ventas_por_mes, labels=ventas_por_mes.index, autopct='%1.1f%%')
plt.title('Ventas por mes')
plt.axis('equal')

# Mostrar el mes con m√°s ventas
print(f"El mes con m√°s ventas fue el {mes_mas_ventas} con {ventas_mes_mas_ventas} ventas.")

# Mostrar la gr√°fica
plt.show()
```

```python
import pandas as pd

# Cargar los datos del archivo Excel
archivo_excel = "/content/shopsales23.xlsx"
datos = pd.read_excel(archivo_excel)

# Calcular el costo unitario de cada producto
datos['costo_unitario'] = datos['total'] / datos['lineitem_quantity']

# Encontrar el producto m√°s caro o los productos m√°s caros
producto_mas_caro = datos.loc[datos['costo_unitario'].idxmax()]

# Imprimir el ID del producto m√°s caro y su costo unitario
print(f"El producto m√°s caro es el ID {producto_mas_caro['lineitem_id']}, con un costo unitario de ${producto_mas_caro['costo_unitario']:.2f}.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos del archivo Excel
archivo_excel = "/content/shopsales23.xlsx"
datos = pd.read_excel(archivo_excel)

# Contar las ventas por estado
ventas_por_estado = datos['shipping_province'].value_counts()

# Obtener el estado con m√°s ventas
estado_mas_ventas = ventas_por_estado.idxmax()
ventas_estado_mas_ventas = ventas_por_estado.max()

# Obtener el estado con menos ventas
estado_menos_ventas = ventas_por_estado.idxmin()
ventas_estado_menos_ventas = ventas_por_estado.min()

# Graficar las ventas por estado
plt.figure(figsize=(10, 6))
ventas_por_estado.plot(kind='bar', color='skyblue')
plt.title('Ventas por Estado')
plt.xlabel('Estado')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Imprimir el estado con m√°s y menos ventas
print(f"El estado con m√°s ventas fue {estado_mas_ventas} con {ventas_estado_mas_ventas} ventas.")
print(f"El estado con menos ventas fue {estado_menos_ventas} con {ventas_estado_menos_ventas} ventas.")

# Mostrar la gr√°fica
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos del archivo Excel
archivo_excel = "/content/shopsales23.xlsx"
datos = pd.read_excel(archivo_excel)

# Contar las ventas canceladas y activas
ventas_canceladas = datos['cancelled_at'].notna().sum()
ventas_activas = datos['cancelled_at'].isna().sum()

# Calcular el porcentaje de ventas canceladas y activas
total_ventas = len(datos)
porcentaje_canceladas = (ventas_canceladas / total_ventas) * 100
porcentaje_activas = (ventas_activas / total_ventas) * 100

# Crear un DataFrame para la gr√°fica de pastel
df = pd.DataFrame({'Estado': ['Canceladas', 'Activas'],
                   'Porcentaje': [porcentaje_canceladas, porcentaje_activas]})

# Graficar el porcentaje de ventas canceladas y activas
plt.figure(figsize=(8, 6))
plt.pie(df['Porcentaje'], labels=df['Estado'], autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'], startangle=140)
plt.title('Porcentaje de Ventas Canceladas vs Activas')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Imprimir el porcentaje de ventas canceladas
print(f"El porcentaje de ventas canceladas es del {porcentaje_canceladas:.2f}%.")
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# Cargar los datos del archivo xlsx
archivo_xlsx = "/content/shopsales23.xlsx"
datos = pd.read_excel(archivo_xlsx)

# Contar las ventas por estado
ventas_por_estado = datos['shipping_province'].value_counts()

# Cargar el archivo shapefile de M√©xico
mexico = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Unir los datos de ventas con el shapefile de M√©xico
mexico_ventas = mexico.merge(ventas_por_estado, how='left', left_on='name', right_index=True)

# Graficar el mapa de calor de M√©xico
fig, ax = plt.subplots(figsize=(10, 8))
mexico_ventas.plot(column='shipping_province', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
plt.title('Mapa de Calor de Ventas por Estado en M√©xico')
plt.axis('off')
plt.show()
```
